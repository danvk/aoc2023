```python
def intersect(px0, vx0, px1, vx1):
    # px0 + t * vx0 = px1 + t * vx1
    # px0 - px1 = t * (vx1 - vx0)
    # t = (px0 - px1) / (vx1 - vx0)
    return (px0 - px1) / (vx1 - vx0)
```
